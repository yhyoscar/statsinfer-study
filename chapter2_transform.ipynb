## 1D Probability Transformation 

```python
import numpy as np
import scipy.stats as stats
import matplotlib 
import matplotlib.pyplot as plt

# uniform -> normal
def fu2n(x, loc=0.0, scale=1.0):
    xmin = np.min(x); xmax = np.max(x)
    percent = (x-xmin)/(xmax-xmin)
    percent[percent<1e-6] = 1e-6
    percent[percent>(1-1e-6)] = 1-1e-6
    return stats.norm.ppf(percent, loc=loc, scale=scale)

x = stats.uniform.rvs(loc=3.0, scale=2.0, size=10000)
y = fu2n(x)

plt.figure(1, figsize=(12, 4))
plt.subplot(1,2,1)
bins = np.linspace(3,5,40)
xpdf = plt.hist(x, bins=bins, density=True)
plt.plot(bins, stats.uniform.pdf(bins, loc=3.0, scale=2.0), 'r')
plt.title('X: uniform')

plt.subplot(1,2,2)
bins = np.linspace(-4, 4, 40)
ypdf = plt.hist(y, bins=bins, density=True)
plt.plot(bins, stats.norm.pdf(bins, loc=0, scale=1.0), 'r')
plt.title('Y: normal')

plt.show()
```

```python
# uniform -> customer (data) defined distribution
def sampling_from_data(data, nsample=10000):
    sdata = sorted(data)
    ndata = len(data)
    perc = stats.uniform.rvs(size=nsample)
    return np.interp(perc, np.arange(ndata)/(ndata-1), sdata)

data = [-1, 4, 0, 2, -1.2, 1.5, 2, 3.5, 0.4, -0.3]

plt.figure(1,figsize=(12, 4))
plt.subplot(1,2,1)
sdata = np.append(sorted(data), [max(data)])
for i in range(len(data)):
    plt.plot([sdata[i]], [(i+1)*1.0/len(data)], 'ro' )
    plt.plot([sdata[i],sdata[i+1]], [(i+1)*1.0/len(data)]*2, 'r-' )

plt.subplot(1,2,2)
xpdf = plt.hist(sampling_from_data(data=data, nsample=10000), 40, density=True) 
```
